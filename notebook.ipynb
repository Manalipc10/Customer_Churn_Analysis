```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

StandardScaler is to Standardize Feature
Mean=0, SD=1

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

joblib.dump(scaler, "scaler.pkl") saves the fitted StandardScaler to a file so you can reuse the same scaling parameters (mean & std) later for new data, ensuring consistent preprocessing during prediction without refitting

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print("Accuracy score on model is {:.3f}".format(accuracy_score(y_test, predictions)))
```

```python
from sklearn.model_selection import GridSearchCV
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
log_model.fit(X_train, y_train)
```

```python
X
```

```python
log_model.predict([[40,0, 20, 80]])
```

```python
y_pred=log_model.predict(X_test)
```

```python
modelperformance(y_pred)
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
param_grid={
    "n_neighbors":[3,5,7,9],
    "weights":["uniform","distance"],
}
```

```python
gridkn=GridSearchCV(KNeighborsClassifier(),param_grid, cv=5)
```

```python
gridkn.fit(X_train,y_train)
```

```python
gridkn.best_params_
```

```python
y_pred=gridkn.predict(X_test)
```

```python
modelperformance(y_pred)
```

svm 

```python
from sklearn.svm import SVC
```

```python
svm=SVC()
```

```python
param_grid ={
    "C":[0.01,0.1,0.5,1],
    "kernel":["linear","rbf","poly"]
}
```

```python
gridsvc=GridSearchCV(svm,param_grid,cv=5)
```

```python
gridsvc.fit(X_train,y_train)
```

```python
gridsvc.best_params_
```

```python
y_pred=gridsvc.predict(X_test)
```

```python
modelperformance(y_pred)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid ={
    "criterion":["gini","entropy"],
    "splitter":["best","random"],
    "max_depth":[None,10,20,30],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
}
```

```python
grid_tree=GridSearchCV(DecisionTreeClassifier(),param_grid,cv=5)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
grid_tree.best_params_
```

```python
y_pred=grid_tree.predict(X_test)
```

```python
y_pred
```

```python
modelperformance(y_pred)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model=RandomForestClassifier()
```

```python
param_grid ={
    "n_estimators":[32,64,128,256],
    "max_features":[2,3,4],
    "bootstrap":[True, False]
}
```

```python
grid_rfc=GridSearchCV(rfc_model,param_grid, cv=5)
```

```python
grid_rfc.fit(X_train,y_train)
```

```python
grid_rfc.best_params_
```

```python
y_pred=grid_rfc.predict(X_test)
```

```python
modelperformance(y_pred)
```

```python
best_model=grid_rfc.best_estimator_
```

```python
joblib.dump(best_model,"model.pkl")
```
